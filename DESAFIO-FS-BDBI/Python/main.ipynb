### **2. Agrupamento de Dados e Plotagem de Gráficos:**

#### a. Plote um gráfico de barras (bar) que exiba a quantidade de funcionários em cada setor (coluna Department);

```python
group_departament = df.groupby('Department').size()
group_departament.plot.barh()

plt.ylabel('')
plt.xlabel("Quantidade de funcionários")
```

- Nota-se que a quantidade de funcionário é maior em "Research & Development", enquanto em "Human Resources" possui a menor quantidade de funcionários da base.

#### b. Plote um gráfico de linha (line) que exiba a quantidade de funcionários para cada nível de escolaridade (coluna Education);

```python
group_departament = df.groupby('Education')['Department'].size()
group_departament.plot.line()

plt.xlabel("Nível de escolaridade")
plt.ylabel("Quantidade de funcionários")
```

- Percebe-se que o nível de escolariedade comum entre os funcionário é no nível 3, enquanto que o nível 5 possui menos funcionários.

#### c. Plote um gráfico de torta/pizza (pie) que exiba a quantidade de funcionários que tem ou não algum atrito na empresa (coluna Attrition).

```python
group_attrition = df.groupby('Attrition').size()
group_attrition.plot.pie(autopct='%1.1f%%')
plt.ylabel("")
```

- Nota-se que 16.1% dos funcionários possui algum atrito na empresa enquantos os 83.9% não possuem.
